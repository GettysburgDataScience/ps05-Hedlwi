### Problem 1

 - How many passengers are recorded in this log?
 - What percentage are male?
 - What percentage survived?

Hint: You can find these three answers with a single command.

````python

total_passengers = data.shape[0]

male_passengers = data[data['Sex'] == 1].shape[0]
percentage_male = (male_passengers/total_passengers)*100

survived_passengers = data[data['Survived'] == 1].shape[0]
percentage_survived = (survived_passengers / total_passengers) * 100

print(total_passengers, percentage_male, percentage_survived)
````

### Problem 2:

````python
'''your code here'''

total_male_passengers = data[data['Sex'] == 1 ].shape[0]

total_female_passengers = data[data['Sex'] == 0 ].shape[0]

male_survived = data[(data['Sex'] == 1 ) & (data['Survived'] == 1)].shape[0]

female_survived = data[(data['Sex'] == 0 ) & (data['Survived'] == 1)].shape[0]

percentage_male_survived = (male_survived / total_male_passengers) * 100

percentage_female_survived = (female_survived / total_female_passengers) * 100

print(f'Percentage of male passengers who survived: {percentage_male_survived:.2f}%')
print(f'Percentage of female passengers who survived: {percentage_female_survived:.2f}%')
````

### Problem 3

Use ```groupby``` and ```agg``` to find the percentage of the total tickets and the survivorship percentage for each class of ticket.

````python
'''your code here'''
class_stats = data.groupby('Pclass').agg(total_tickets=('PassengerId', 'count'),survivors=('Survived', 'sum'))

class_stats['percentage_total_tickets'] = (class_stats['total_tickets'] / total_passengers) * 100

class_stats['percentage_survived'] = (class_stats['survivors'] / class_stats['total_tickets']) * 100

print(class_stats[['percentage_total_tickets', 'percentage_survived']])
````

### Problem 5

Now that we've inspected the data, let's build a classifier (you'll compare two models) that predicts which passengers survived. These are the steps you should take:

 - Choose your features. The label will be survivorship.
 - Split the data into training and test sets.
 - Standardize the values using StandardScaler
 - Choose two of the following model types: KNN, SVM, Decision Tree. Create those models and select hyper-parameters.
 - Fit your models to the training data.
 - Apply the fit models to create predictions from the test data.
 - For each model, create a confusion matrix for the training and test data. You may want to revisit the hyper-parameters and re-fit the models.

 - Answer the questions at the bottom of the code block.

````python
'''Your code here'''
label = 'Survived'
features = ['Sex', 'Age', 'Pclass', 'Fare']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data[features], data[label], test_size=0.2)
````

````python
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

tree_clf = DecisionTreeClassifier(max_depth=4)
tree_clf.fit(X_train, y_train)

fig, ax = plt.subplots(1, 1, figsize=(15, 6))
plot_tree(tree_clf, feature_names=features, class_names=['drowned', 'survived'], filled=True, fontsize=8, ax=ax)

plt.show()

````

````python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

y_tree_train = tree_clf.predict(X_train)
y_tree_test = tree_clf.predict(X_test)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ConfusionMatrixDisplay.from_predictions(y_train, y_tree_train, normalize='true', display_labels=['drowned', 'survived'], ax=ax[0])
ConfusionMatrixDisplay.from_predictions(y_test, y_tree_test, normalize='true', display_labels=['drowned', 'survived'], ax=ax[1])

ax[0].set_title('Training')
ax[1].set_title('Testing')

plt.show()
````

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(X_train)

Xs_train = ss.transform(X_train)
Xs_test = ss.transform(X_test)

ys_train = y_train.values.ravel()
ys_test = y_test.values.ravel()

knn_clf = KNeighborsClassifier(n_neighbors = 15)
knn.fits = (Xs_train, ys_train)
````
